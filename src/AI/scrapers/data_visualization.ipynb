```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/combined_team_player_data.csv')
```

```python
labels=['scored', 'no-score']

scored = df[df['ppPoints'] >= 1]['ppPoints']
no_score = df[df['ppPoints'] == 0]['ppPoints']

concat = pd.concat([scored, no_score])

sns.countplot(x=concat, data=df, palette='hls')
print('Players who scored: ', len(scored))
print('Players who did not score: ', len(no_score))

perc = (scored.shape[0] / df['ppPoints'].shape[0]) * 100
print(f'Percentage of players with a point ({scored.shape[0]}/{df["ppPoints"].shape[0]}): {round(perc, 2)}%')
```

### Reduce down to binary classes
We want to know if the player scores a powerplay point or not. It's quite rare for a player to score 2 or more (although it may be important information)

```python
df['ppPoints'] = df['ppPoints'].map(lambda x: 1 if x >= 1 else 0)
```

```python
labels=['scored', 'no-score']

scored = df[df['ppPoints'] >= 1]['ppPoints']
no_score = df[df['ppPoints'] == 0]['ppPoints']

concat = pd.concat([scored, no_score])

sns.countplot(x=concat, data=df, palette='hls')
print('Players who scored: ', len(scored))
print('Players who did not score: ', len(no_score))

perc = (scored.shape[0] / df['ppPoints'].shape[0]) * 100
print(f'Percentage of players with a point ({scored.shape[0]}/{df["ppPoints"].shape[0]}): {round(perc, 2)}%')
```
